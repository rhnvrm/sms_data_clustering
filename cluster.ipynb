```python
import numpy as np
import pandas as pd
import random
```

```python
sim = np.load("sim.npy")
size = len(sim)
```

```python
corpus = pd.DataFrame.from_csv('corpus.csv')
corpus = corpus[:size]
corpus["cluster"] = corpus.apply(lambda x: -1, axis=1)
corpus = corpus.reset_index()
corpus["orig_index"] = corpus.apply(lambda x: x["index"], axis=1)
corpus
```

```python
thresh = 0.0007
cluster_num = 0

cluster = {}

s = list(range(0,size))

while(s):
    cluster[cluster_num] = [s.pop(0)]
    for i in s:
        #print(sim[i][cluster[cluster_num]])
        if sim[i][cluster[cluster_num]][0] >= thresh:
            cluster[cluster_num] += [i]
            s.remove(i)
    for i in cluster[cluster_num]:
        for j in s:
            if sim[j][i] >= thresh:
                cluster[cluster_num] += [j]
                s.remove(j)
    cluster_num += 1

cluster
```
